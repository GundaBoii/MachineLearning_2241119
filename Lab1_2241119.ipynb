```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from math import pi

# Update the file path to match the location of your dataset
file_path = 'songs_normalize.csv'  # Use the full path if the file is not in the same directory

# Load the dataset
df = pd.read_csv(file_path)
```

```python
# Check the shape of the dataset
print("Dataset shape:", df.shape)

# Display column names
print("Columns in the dataset:", df.columns.tolist())

# Check for missing values
print("\nMissing values per column:\n", df.isnull().sum())

# Data types of columns
print("\nData types:\n", df.dtypes)

# Summary statistics
df.describe(include='all')
```

```python
#Univariate Analysis

# Top 10 Artists With Most Songs
top_artists = df['artist'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_artists.values, y=top_artists.index, palette='viridis')
plt.title('Top 10 Artists with Most Songs')
plt.xlabel('Number of Songs')
plt.ylabel('Artist Name')
plt.show()
```

```python
###
# Histogram Findings
# The bar chart highlights the artists with the highest number of onsgs in the dataset.
# It reflects the prominence of these artists in the music industry.
###
```

```python
#Bivariate Analysis
#Relationship b/w Danceability & Energy

plt.figure(figsize=(8, 5))
sns.scatterplot(x='danceability', y='energy', hue='popularity', size='popularity', data=df, palette='coolwarm')
plt.title('Danceability vs Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.show()
```

```python
###
# Bivariate Analysis Findings
# The scatterplot shows how danceability and energy correlate, with point size and color indicating
# popularity. Songs with high danceability and energy tend to have higher popularity
# It also highlights listener preferences for upbeat tracks.
###
```

```python
# Multivariate Analysis
# Select a random song
example_song = df.sample(1).iloc[0]
features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'valence']

# Prepare data for radar chart
values = [example_song[feature] for feature in features]
values += values[:1]  # Close the circle
angles = [n / float(len(features)) * 2 * pi for n in range(len(features))]
angles += angles[:1]

# Plot radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='skyblue', alpha=0.4)
ax.plot(angles, values, color='blue', linewidth=2)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(features)
plt.title("Audio Features Radar Chart")
plt.show()
```

```python
###
# Plot Radar Chart Findings
# The radar plot visualizes audio features of a single song, providing a holistic view of its
# characterisitcs like danceability, energy and acoustiness.
# Such charts help compare songs or genres.
###
```
