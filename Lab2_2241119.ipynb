```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from math import pi

# Update the file path to match the location of your dataset
file_path = 'songs_normalize.csv'  # Use the full path if the file is not in the same directory

# Load the dataset
df = pd.read_csv(file_path)
```

```python
###
# Categorical Data Analysis
# Identifying All Categorical Columns In Dataset
###

categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical Columns:\n", categorical_cols)

###
# Categorical Columns Are Identified
###
```

```python
###
# Count Plot For Univariate Column
###

plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='genre', order=df['genre'].value_counts().index, palette='viridis')
plt.title('Count of Songs by Genre')
plt.show()

# After Doing Univariate Analysis, The Chart Shows Most & Least Represented Genres
```

```python
###
# Count Plot For Bivariate Analysis
###

# Average popularity by genre
genre_popularity = df.groupby('genre')['popularity'].mean().sort_values(ascending=False)

# Bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_popularity.values, y=genre_popularity.index, palette='coolwarm')
plt.title('Average Popularity by Genre')
plt.xlabel('Average Popularity')
plt.ylabel('Genre')
plt.show()

# After Doing Bivariate Analysis, The Chart Shows Listener Prefereces
```

```python
###
# Top 10 Artists By Song Count
top_artists = df['artist'].value_counts().head(10)

# Bar Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=top_artists.values, y=top_artists.index, palette='spring')
plt.title('Top 10 Artists by Song Count')
plt.xlabel('Number of Songs')
plt.ylabel('Artist')
plt.show()

# After Doing Bivariate Analysis, The Chart Shows The Top 10 Artists By Songs Released
```

```python
###
# Heatmap For Correlation And Energy & Danceability
correlation = df[['energy', 'danceability']].corr()

# Plotting Heatmao
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap Between Energy and Danceability')
plt.show()

# After Doing Bivariate Analysis, The Chart Shows The Top 10 Artists By Songs Released
```

```python
# Violin Plot For Danceability By Genre

plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='danceability', y='genre', inner='quartile', palette='pastel')
plt.title('Danceability Distribution by Genre')
plt.xlabel('Danceability')
plt.ylabel('Genre')
plt.show()

# Danceability Has Wider Variability In Some Genres Like Pop, While It’s More Uniform In Classical
```
