```python
#Lab 3 - Numerical & Categorical EDA

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from math import pi

# File Path
file_path = 'deliveries.csv'

# Load Dataset
df = pd.read_csv(file_path)
```

```python
#Numerical EDA

# Plots Used:
# 1. Boxplot
# 2. Scatter Plot
# 3. Line Plot
# 4. Correlation Heatmap
```

```python
# Sample data to reduce overpopulation
sampled_data = df.sample(frac=0.3, random_state=42) #0.3 indicates using only 30% of the actual data

sns.histplot(sampled_data['total_runs'], kde=True, bins=20, color='blue')
plt.title('Distribution of Total Runs (Sampled Data)')
plt.xlabel('Total Runs')
plt.ylabel('Frequency')
plt.show()

#This histogram shows the distribution of total runs scored in individual deliveries during cricket matches.
```

```python
#Boxplot For Total Runs

sns.boxplot(x=sampled_data['batsman_runs'], color='orange')
plt.title('Boxplot of Batsman Runs (Sampled Data)')
plt.xlabel('Batsman Runs')
plt.show()

#The boxplot displays the spread of individual batsman runs. 
#The middle line shows the median runs per delivery, while the upper/lower bounds of the box indicate the interquartile range.
```

```python
#Scatter Plot For Overs & Total Runs

scatter_sample = df.sample(500, random_state=42)  # Reduce to 500 records
sns.scatterplot(x=scatter_sample['over'], y=scatter_sample['total_runs'], color='green')
plt.title('Overs vs. Total Runs (Sampled Data)')
plt.xlabel('Over')
plt.ylabel('Total Runs')
plt.show()

#This scatter plot shows the relationship between overs and runs scored.
```

```python
#Line Plots For Avg Runs

avg_runs_per_ball = df.groupby('ball')['total_runs'].mean().reset_index()
sns.lineplot(x=avg_runs_per_ball['ball'], y=avg_runs_per_ball['total_runs'], color='purple')
plt.title('Average Total Runs Over Balls')
plt.xlabel('Ball')
plt.ylabel('Average Total Runs')
plt.show()

#This line plot shows how scoring trends evolve ball-by-ball.
#It highlights specific balls where runs scored may tend to increase due to strategy or pressure situations.
```

```python
#Correlation Heatmap For Selected Columns

selected_numerical_cols = df[['over', 'ball', 'total_runs', 'batsman_runs']]
plt.figure(figsize=(8, 6))
sns.heatmap(selected_numerical_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Selected Numerical Variables')
plt.show()

#The heatmap displays correlations between numerical features like overs, balls, total runs, and batsman runs.
```

```python
#Categorical EDA

#Plots Used:
#1. Bar Plot [Batting, Bowling & Dismissal Types]
#2. Stacked Bar Plot
#3. Pie Chart
```

```python
#Bar Plot For Top Batting Teams

top_teams = df['batting_team'].value_counts().head(10)
sns.barplot(y=top_teams.index, x=top_teams.values, palette='viridis')
plt.title('Top 10 Batting Teams by Frequency')
plt.xlabel('Frequency')
plt.ylabel('Batting Team')
plt.show()

#This bar plot displays the most frequent batting teams in the dataset.
#It reveals which teams play the most matches and therefore dominate the dataset in terms of appearances.
```

```python
#Bar Plot For Top Bowling Teams

top_bowling_teams = df['bowling_team'].value_counts().head(10)
sns.barplot(y=top_bowling_teams.index, x=top_bowling_teams.values, palette='cubehelix')
plt.title('Top 10 Bowling Teams by Frequency')
plt.xlabel('Frequency')
plt.ylabel('Bowling Team')
plt.show()

#This plot shows the most frequent bowling teams.
#Like the batting teams, it highlights which teams bowl the most often and provides insights into team activity.
```

```python
#Stacked Bar Plot

extra_runs_by_team = df.groupby('batting_team')['extra_runs'].sum().sort_values().tail(10)
extra_runs_by_team.plot(kind='barh', color='teal')
plt.title('Top 10 Teams by Total Extra Runs')
plt.xlabel('Extra Runs')
plt.ylabel('Batting Team')
plt.show()

#This horizontal bar plot highlights the teams that gain the most extra runs, which may result from opponents' mistakes like wides or no-balls.
#It reflects which teams capitalize on such errors.
```

```python
#Bar Plot For Dismissal Types 

dismissal_counts = df['dismissal_kind'].value_counts().head(5)
sns.barplot(y=dismissal_counts.index, x=dismissal_counts.values, palette='coolwarm')
plt.title('Top 5 Types of Dismissals')
plt.xlabel('Frequency')
plt.ylabel('Dismissal Kind')
plt.show()

#This bar plot shows the most common ways batsmen get out.
```

```python
#Pie Chart For Super Overs

df['is_super_over'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'salmon'], labels=['No', 'Yes'])
plt.title('Percentage of Matches with Super Overs')
plt.ylabel('')
plt.show()

#The pie chart indicates how often super overs occur in matches.
#A small proportion of "Yes" highlights their rarity, which may happen only in tied matches.
```
